## Data Encoding and Exploration
This data needs to be encoded so that it can be used. My first approach is the following:

pclass will be encoded using the one hot encoding methodology, where we will replace this column with 3 columns of binary inputs representing each class. An input will be 1 if it is true (i.e. the passanger is in the corresponding class) and 0 if it is false. This method was chosen in place of a simple {-1 = '1st', 0 = '2nd', 1='3rd'} to allow for weightings to be non-linear. 

sex will be encoded {1 = 'male, -1 = 'female'}

Age, Sibsp, Parch will be normalised. 

Ticket will need to be formatted as some data points are integers, whilst some include non-numeric characters. Following that, we could attempt normalisation however will need to get a better understanding for what the data is.

Fare will be normalised.

Cabin may be removed, or tackled at a later date.

The embarked column will be encoded in a similar format to the pclass column, where we will use one hot encoding to replace this column with 3 columns of binary inputs representing port of embarkation. 


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re
from encode import *

from pandas import Series, DataFrame


filename = 'train.csv'
df = pd.read_csv(filename)
df.columns = ['PassengerId', 'Survived', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked']

passengers = df
print(passengers.info)

for column in ['Survived', 'Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked']:
	print(f'Column "{column}" contains the following unique inputs: {passengers[column].unique()}')

print('\nTable of missing value counts in each column:')
print(passengers.isnull().sum())
```

We have some issues with missing data in the Age, Cabin and Embarked columns. We will need to fix this before continuing.

We will remove the Cabin column to begin with, as at this stage of analysis, I do not know enough about the cabin entries and we are missing such a large amount of data. It is possible that we could encode using a binary input, whether they have a cabin or not. I will consider this on the second iteration of this model.

The Embarked column will be adjusted by filling in null values with a 0, representing no port information. (After encoding, this will be presented by a 0 in all three columns).

As for the Age column, I will begin by filling it with the mean age value. 

```python
# Sorting out missing data
passengers['Embarked'] = passengers['Embarked'].fillna(0)
passengers = passengers.drop(columns=['Cabin'])
passengers['Age'] = passengers['Age'].fillna(passengers['Age'].mean())
```

Now we will begin encoding the data.

```python
def encode_pclass(df):
	# A function used to create three columns, encoding the pclass column
	class1, class2, class3= [], [], []
	for passenger in df['Pclass']:
		if passenger == 1:
			class1.append(1)
			class2.append(0)
			class3.append(0)
		elif passenger == 2:
			class1.append(0)
			class2.append(1)
			class3.append(0)
		elif passenger == 3:
			class1.append(0)
			class2.append(0)
			class3.append(1)
		else: 
			class1.append(0)
			class2.append(0)
			class3.append(0)
	df['class1'] = class1
	df['class2'] = class2
	df['class3'] = class3
	print(f'pclass variable encoding complete. Columns "class1", "class2" and "class3" added to DataFrame')
	print(f'IMPORANT: ensure you drop the original pclass column.\n')


def encode_embarked(df):
	# A function used to create three columns, encoding the Embarked column
	cherbourg, queenstown, southampton = [], [], []
	for passenger in df['Embarked']:
		if passenger == 'C':
			cherbourg.append(1)
			queenstown.append(0)
			southampton.append(0)
		elif passenger == 'Q':
			cherbourg.append(0)
			queenstown.append(1)
			southampton.append(0)		
		elif passenger == 'S':
			cherbourg.append(0)
			queenstown.append(0)
			southampton.append(1)
		else:
			cherbourg.append(0)
			queenstown.append(0)
			southampton.append(0)
	df['cherbourg'] = cherbourg
	df['queenstown'] = queenstown
	df['southampton'] = southampton
	print(f'Embarked variable encoding complete. Columns "cherbourg", "queenstown" and "southampton" added to DataFrame')
	print(f'IMPORANT: ensure you drop the original Embarked column.\n')

def encode_sex(df):
	# A function used to replace the sex column in a df, with a binary value. 
	sex = []
	for passenger in df['Sex']:
		if passenger == 'male':
			sex.append(1)
		else:
			sex.append(-1)
	df['Sex']= sex
	print(f'Sex variable encoding complete in DataFrame.\n')


def encode_ticket(df):
	# A function used to remove the letters from the ticket column.
	df['Ticket'] = df['Ticket'].str.extract('(\d+)(?!.*\d)')
	df['Ticket'] = df['Ticket'].astype(float)
	print(f'Non-numeric characters successfully removed from ticket column.\n')


def normalise_col(df, col):
	# Function to normalise the data in each column.
	mean, std  = df[col].mean(), df[col].std()
	normalised_col = (df[col] - df[col].mean())/(df[col].std())
	df[col] = normalised_col
	print(f'Column normalised in DataFrame.')
	return mean, std

encode_pclass(passengers)
encode_embarked(passengers)
encode_sex(passengers)
encode_ticket(passengers)
passengers['Ticket'] = passengers['Ticket'].fillna(passengers['Ticket'].mean())
passengers = passengers.drop(columns=['Pclass', 'Embarked'])
```

## Model Implementation

I have chosen to use a logistic regression model for this ML problem. 

Firstly, it is important to remove some of the existing columns. We will remove the Name, PassengerId, class1 and cherborg columns. The Name column will be removed as it is non-numeric. The PassengerId column is being removed because the sample is expected to be taken randomly, as such, the ID number of the data set should have no effect on survival chance. Class1 and cherborg columns are being removed due to the linearly dependant nature of the class and embarkation columns.
 
We will also add a 'bias' column to the data set, which represents the alpha intercept. This will be a column of 1's.

We will use statsmodels for feature selection. We will not use regularisation for this part, however will apply regularisation for the actual implementation, using scikit learn. We will then evaluate the model against the test data set and adjust the model accordingly. (Spltting the dataset into  70:30, train:test sets).

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split

passengers_final = passengers.drop(columns=['Name', 'PassengerId','class1', 'cherbourg']).copy()
passengers_final['bias'] = np.ones(passengers_final.shape[0]) # Adding the bias column

X = passengers_final.loc[:, passengers_final.columns != 'Survived']
y = passengers_final.loc[:, passengers_final.columns == 'Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

logit_model = sm.Logit(y_train.astype(int),X_train.astype(float))

result = logit_model.fit()
result.summary2()
```

Evaluating the summary output, we note that the following features have large p-values:
- Parch
- Ticket
- Fare
- queenstown
- southampton

This suggests we should consider removing these features from the data set as their beta coefficient is close to 0.

Let's look at the effectivity of this model on our test data set.

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import confusion_matrix

logreg = LogisticRegression(fit_intercept=False)
logreg.fit(X_train.astype(float), np.ravel(y_train))

y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))

cm = confusion_matrix(y_test, y_pred)
print('\n', cm)

print('\n',logreg.coef_)
```

As we can see, this model has difficulty predicting if a person will actually survive. We have 65 times where a  person was predicted to die however they in fact survived. Let's try to improve the accuracy of the model. 

To do so, we will consider also removing the Ticket, queenstown, Fare and Parch columns.
I have chosen to remove these columns due to their p-values being relatively high. I have chosen not to remove southampton on this iteration as I would like to investigate to see whether there is an effect on port of embarkation on the chance of survival. 

```python
passengers_final = passengers.drop(columns=['Name', 'PassengerId','class1', 'cherbourg', 'Ticket', 'queenstown', 'Fare', 'Parch']).copy()
passengers_final['bias'] = np.ones(passengers_final.shape[0]) # Adding the bias column

X = passengers_final.loc[:, passengers_final.columns != 'Survived']
y = passengers_final.loc[:, passengers_final.columns == 'Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

logit_model = sm.Logit(y_train.astype(int),X_train.astype(float))

result = logit_model.fit()
result.summary2()
```

In our new model, we see that all of the variables have a p-value lower than 0.05. As a result, there is currently no evidence to suggest that these variables have no effect on the survival chances of a passenger. 

```python
logreg = LogisticRegression(fit_intercept=False)
logreg.fit(X_train.astype(float), np.ravel(y_train))

y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))

cm = confusion_matrix(y_test, y_pred)
print('\n', cm)

print('\n',logreg.coef_)
```

As you can see, we have significantly improved the model. We have improved its accuracy by 7%, and improved its rate of correct classificaiton of people who survive. On the other hand, we have doubled the amount of survived passengers classified as dead. A deeper breakdown of the model accuracy can be seen below. 

```python
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve

print(classification_report(y_test, y_pred))

logit_roc_auc = roc_auc_score(y_test, logreg.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test.values, logreg.predict_proba(X_test)[:,1])

plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```
